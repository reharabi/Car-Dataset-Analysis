##**🧹 Data Cleaning**

1. Handling Missing Data:
* Identify missing values
* How to handle missing data for each column
   * Replace with "missing data"
   * Replace with median
   * Delete rows

2. Data Type Conversion:

   Ensure that columns are of the correct data type for analysis

3. Filter Data:

   Filter to only include cars from 1995 and later

4. String Operations:
   Standarize these columns' entries(Vehicle Style, Market Category) to lowercase

5. Handling Duplications:
* Identify duplicated rows
* How to handle duplicated rows
  * Delete duplicated rows


```python
# 1 : Handling Missing Data:
# Identify missing values

print("\nMissing Values:")
print(car_df.isnull().sum()) # total missing values per column
```

```python
# Decide how to handle missing data for each column

car_df['Market Category'] = car_df['Market Category'].fillna('missing data') # replace with "missing data" due to high amount,needs to be checked
car_df.fillna({'Engine HP': car_df['Engine HP'].median()}, inplace=True)#replace with median
car_df.fillna({'Engine Cylinders': car_df['Engine Cylinders'].median()}, inplace=True)#replace with median
car_df.dropna(subset=['Number of Doors','Engine Fuel Type'], inplace=True)#delete rows which has low amount

print(car_df.isnull().sum()) #check for null values in all columns after handling missing data
```

```python
# 2 : Data Type Conversion:
# Ensure that columns are of the correct data type for analysis
# result : all correct

print(car_df.dtypes)
```

```python
# 3 : Filter Data:
# Filter to only include cars from the 1995 and later

filtered_car_df = car_df[car_df['Year']>= 1995]
filtered_car_df.head(10)
```

```python
# 4 : String Operations:
# Standardize text entries by converting these columns' entries to lowercase

#filtered_car_df['Vehicle Style'] = filtered_car_df['Vehicle Style'].str.lower()       # change selected columns to lowercase
#filtered_car_df['Market Category'] = filtered_car_df['Market Category'].str.lower()
#filtered_car_df

filtered_car_df = filtered_car_df.map(lambda x: x.lower() if isinstance(x, str) else x) # change all string values to lowercase
filtered_car_df.head(10)
```

```python
# 5. Handling Duplications:
# Identify duplicated rows

filtered_car_df[filtered_car_df.duplicated()]

```

```python
# How to handle duplicated rows

filtered_car_df.drop_duplicates(inplace=True) # delete duplicated rows
print("\nDuplicate Rows:", filtered_car_df.duplicated().sum())
```

##**🛠 Feature Engineering**

1. Create New Columns:
 * Create a column called "Total MPG" that is the average of "city mpg" and "highway MPG"
 * Create a column called "Price per HP" calculated as(MSRP/Engine HP)

```python
# 1 : Create New Columns:
# Create a column called Total MPG that is the average of city mpg and highway MPG

filtered_car_df['Total MPG'] = (filtered_car_df['city mpg'] + filtered_car_df['highway MPG'])/2
print(filtered_car_df['Total MPG'])
```

```python
# Create a column called Price per HP calculated as: MSRP / Engine HP

filtered_car_df['Price per HP'] = filtered_car_df['MSRP'] / filtered_car_df['Engine HP']
filtered_car_df['Price per HP'] = filtered_car_df['Price per HP'].round(2)
print(filtered_car_df['Price per HP'])
```

```python
# 2 : Group Analysis:

def group_and_mean(group_col, mean1_col, mean2_col):
    return filtered_car_df.groupby(group_col).agg({mean1_col: 'mean', mean2_col: 'mean'}).round(2) # define function to avoid code repetition
```

```python
# Group the data by following columns and calculate the average MSRP and Popularity for each group :Driven_Wheels, Vehicle Size & Engine Cylinders

driven_wheels_group = group_and_mean('Driven_Wheels', 'MSRP', 'Popularity')
print(driven_wheels_group)
```

```python
vehicle_size_group = group_and_mean('Vehicle Size', 'MSRP', 'Popularity')
print(vehicle_size_group)
```

```python
engine_cylinders_group = group_and_mean('Engine Cylinders', 'MSRP', 'Popularity')
print(engine_cylinders_group)
```

```python
# Q3 : Visualizations:

# A histogram that shows a distribution for the city mpg column

plt.figure(figsize=(12, 8))
plt.hist(filtered_car_df['city mpg'], bins=20, edgecolor='saddlebrown', color='olive')
plt.xlabel('City MPG')
plt.ylabel('Frequency of values')
plt.title('Distribution of City MPG')
plt.grid(True)
plt.show()

```

```python
# A boxplot showing the distribution of MSRP for each category in Driven_Wheels

plt.figure(figsize=(12, 8))
sns.boxplot(data= filtered_car_df, x='Driven_Wheels', y='MSRP', color='rosybrown')
plt.title(f'Box Plot of MSRP by Driven_Wheels')
plt.xlabel('Driven_Wheels')
plt.ylabel('MSRP')
plt.title('Distribution of MSRP for each category in Driven_Wheels')
plt.grid(True)
plt.ylim(0, 300000)
plt.show()
```

**Is there a significant difference in price(MSRP) for different drivetrains(Driven_Wheels)?**

1. Significant difference in MSRP across drivetrain types

2. Clear pricing hierarchy based on drivetrain:
* Rear Wheel Drive,most common in luxury and high performance vehicles, have the highest MSRP distribution
* All Wheel Drive, most common in SUV and crossover vehicles, also show relatively high MSRP range
* Four Wheel Drive, often found in utility vehicles and trucks, vary in price but doesn't reach rear wheel and all wheel drive levels
* Front wheel drive, typically economy and compact vehicles, have the lowest MSRP


```python
def find_outliers(group):
    Q1 = group['MSRP'].quantile(0.25)
    Q3 = group['MSRP'].quantile(0.75)
    IQR = Q3 - Q1
    outliers = group[group['MSRP'] > Q3 + 1.5 * IQR]
    return outliers

outlier_df = filtered_car_df.groupby('Driven_Wheels').apply(find_outliers).reset_index(drop=True)

outlier_df[['Make','Model','Market Category','Driven_Wheels','MSRP']].sort_values(by='MSRP', ascending=False) # indentify outliers
```

```python
# A line plot showing trends in average city mpg and highway MPG for different Transmission Type

avg_mpg = filtered_car_df.groupby('Transmission Type')[['city mpg', 'highway MPG']].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(avg_mpg['Transmission Type'], avg_mpg['city mpg'], marker='o', label='Average city MPG')
plt.plot(avg_mpg['Transmission Type'], avg_mpg['highway MPG'], marker='s', label='Average Highway MPG')

plt.title('Average City and Highway MPG by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Miles Per Gallon (MPG)')
plt.xticks(rotation=360)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
custom_palette = ['rosybrown', 'olive', 'saddlebrown', 'darkorange', 'grey']

plt.figure(figsize=(10,6))
sns.scatterplot(data=filtered_car_df, x='city mpg', y='highway MPG', hue='Transmission Type', palette=custom_palette)
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.title('City MPG vs. Highway MPG by Transmission Type')
plt.grid(True)
plt.show()
```

**How do MPG(city mpg and highway MPG) trends change with transmission type(Transmission Type)?**

1. Efficiency advantage of direct-drive. The high spike in mpg for direct-drive highlights the superior fuel economy of electric drivetrains, especially in urban settings
2. Conventional transmissions have relatively similar mpg across. This suggests that those transmissions alone doesn't drastically affect fuel economy, other factors like engine size and weight play a role too
3. Consistent gap between city and highway mpg suggests that vehicles perform more efficiently on highways, regardless of transmission, except for direct-drive


```python
# 4. Correlation Analysis:
# Investigate the correlation between the following variables: Engine HP, MSRP, Popularity, city mpg, and highway MPG

var_data = filtered_car_df.loc[:,['Engine HP','MSRP','Popularity','city mpg','highway MPG']]
correlation_matrix = var_data.corr(method='pearson')

custom_palette = ['rosybrown', 'olive', 'rosybrown', 'olive', 'saddlebrown']
#sns.set_palette(custom_palette) # This line is not needed when passing the palette directly to cmap

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap=custom_palette, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

**Correlation summary**
* Engine HP & MSRP
  * Strong positive correlation. More powerful engines tend to be more expensive
* City MPG & Highway MPG
  * Very strong positive correlation. Vehicles efficient in the city are also efficient on highways
* Engine HP & City MPG
  * Moderate positive correlation. Higher horsepower may slightly improve city mpg, but not strongly
* Engine HP & Highway MPG
  * Similar moderate correlation. Again, not a dominant factor
* MSRP & City MPG
  * Weak negative correlation. More expensive cars tend to have lower city mpg
* MSRP & Highway MPG
  * Slightly stronger negative correlation. Luxury or performance vehicles may have less fuel efficiency
* Popularity & Others
  * Popularity doesn't correlate meaningfully with any other variable. This is likely influenced by brand, marketing, or subjective factors
